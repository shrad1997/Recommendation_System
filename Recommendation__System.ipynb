```python
import pandas as pd
import numpy as np
```

```python
df_book=pd.read_csv('/content/book (1).csv', encoding='latin-1')
```

```python
df_book.head(5)
```

```python
df_book.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df_book.info()
```

```python
df_book.rename(columns={'User.ID':'User_ID','Book.Title':'Book_Title','Book.Rating':'Book_Rating'},inplace=True)
```

```python
df_book.head(5)
```

```python
len(df_book.User_ID.unique())
```

```python
len(df_book.Book_Title.unique())
```

```python
book = df_book.reset_index().pivot_table(index='User_ID',
                                 columns='Book_Title',
                                 values='Book_Rating').reset_index(drop=True)
```

```python
book.shape
```

```python
book
```

```python
# Replacing the index values by unique user Ids
book.index=df_book['User_ID'].unique()
book
```

```python
#Impute those NaNs with 0 values
book.fillna(0, inplace=True)
```

```python
book.head(5)
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(book.values,metric='cosine')
```

```python
user_sim
```

```python
similar_user_df=pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
similar_user_df.index =book.index.unique()
similar_user_df.columns =book.index.unique()
```

```python
similar_user_df.iloc[0:5,0:5]
```

```python
# for Nullyfying diagonal values
np.fill_diagonal(user_sim, 0)
similar_user_df.iloc[0:5,0:5]
```

```python
#Most Similar Users
similar_user_df.idxmax(axis=1)[0:30]
```
